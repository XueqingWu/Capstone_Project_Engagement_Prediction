## Create lag features

```python
# Ensure data is sorted correctly
# weekly_data_by_account = weekly_data_by_account.sort_values(by=["profile_id", "year_week"]).reset_index(drop=True)

# Define lags and relevant engagement-related columns
lags = [1, 2, 3, 4]
lag_features = ["engagement", "followers", "reach", "impressions"]

# Convert 'year_week' into separate numeric year and week for faster processing
weekly_data_by_account[['year', 'week']] = weekly_data_by_account['year_week'].str.split('-', expand=True).astype(int)

# Create a shifted 'year_week' column for detecting consecutive weeks
weekly_data_by_account['prev_year_week'] = weekly_data_by_account.groupby('profile_id')['year_week'].shift(1)

# Function to check if the previous week is exactly lag weeks before
def is_exact_lag_vectorized(df, lag):
    prev_week = df.groupby('profile_id')['week'].shift(lag)
    prev_year = df.groupby('profile_id')['year'].shift(lag)

    # Adjust for year transition
    expected_week = df['week'] - lag
    expected_year = df['year']

    # Handle cases where week becomes negative (previous year)
    expected_year[expected_week <= 0] -= 1
    expected_week[expected_week <= 0] += 52  # Assuming a 52-week year

    return (prev_year == expected_year) & (prev_week == expected_week)

# Create lag features with vectorized checks
for lag in lags:
    for feature in lag_features:
        if feature in weekly_data_by_account.columns:
            lag_col_name = f"{feature}_lag_{lag}"
            weekly_data_by_account[lag_col_name] = weekly_data_by_account.groupby("profile_id")[feature].shift(lag)

            # Set lag values to 0 if the previous week is not consecutive
            weekly_data_by_account.loc[~is_exact_lag_vectorized(weekly_data_by_account, lag), lag_col_name] = 0
```

## Train separate models and save them

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import joblib  # For saving models

# Define target variables
targets = ["engagement_future_7d", "engagement_future_14d", "engagement_future_21d", "engagement_future_28d"]

# Dictionary to store models and results
models = {}
metrics = {}

# Train separate models
for target in targets:
    model = LinearRegression()
    model.fit(X_train, y_train[target])  # Train on each target separately
    models[target] = model  # Store trained model
    
    # Save model to file
    joblib.dump(model, f"{target}_model.pkl")

print("All models trained and saved successfully.")
```

```python
# Dictionary to store predictions
predictions = {}

# Evaluate each model
for target in targets:
    model = models[target]
    
    # Make predictions
    y_pred = model.predict(X_test)
    predictions[target] = y_pred  # Store predictions

    # Calculate evaluation metrics
    mae = mean_absolute_error(y_test[target], y_pred)
    mse = mean_squared_error(y_test[target], y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test[target], y_pred)

    # Store metrics
    metrics[target] = {"MAE": mae, "MSE": mse, "RMSE": rmse, "R²": r2}

    # Print results
    print(f"\nEvaluation for {target}:")
    print(f"MAE: {mae:.4f}")
    print(f"MSE: {mse:.4f}")
    print(f"RMSE: {rmse:.4f}")
    print(f"R²: {r2:.4f}")

print("\nAll predictions and evaluations completed.")
```
