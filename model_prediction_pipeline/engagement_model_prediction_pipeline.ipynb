```python
# Convert 'date' to datetime format and set it as the index
data["date"] = pd.to_datetime(data["date"])
data.set_index("date", inplace=True)

# Group by 'profile_id' and then resample by week and calculate the sum for each group
weekly_data_by_account = data.groupby("profile_id").resample("W").sum()

weekly_data_by_account.drop(columns=["profile_id"], inplace=True)

# Reset the index to make 'date' and 'profile_id' columns again
weekly_data_by_account.reset_index(inplace=True)

# Create a new column to indicate the week number from the 'date' column
weekly_data_by_account["week_of_year"] = (
    weekly_data_by_account["date"].dt.isocalendar().week
)

# Optional: Format the week as 'Year-Week'
weekly_data_by_account["year_week"] = weekly_data_by_account["date"].dt.strftime(
    "%Y-%U"
)

# Sort the data by 'profile_id' and 'year_week'
weekly_data_by_account.sort_values(
    by=["profile_id", "year_week"], ascending=[True, True], inplace=True
)
```

```python
# convert "engagement" and "posts" to integer
weekly_data_by_account["engagement"] = weekly_data_by_account["engagement"].astype(int)
weekly_data_by_account["posts"] = weekly_data_by_account["posts"].astype(int)
# Calculate the engagegment per post if the post is not 0
weekly_data_by_account["engagement_per_post"] = (
    weekly_data_by_account["engagement"] / weekly_data_by_account["posts"]
)
```

## Create lag features

```python
# Define lags and relevant engagement-related columns
lags = [1, 2, 3, 4]
lag_features = ["engagement", "followers", "reach", "impressions"]

# Convert 'year_week' into separate numeric year and week for faster processing
weekly_data_by_account[["year", "week"]] = (
    weekly_data_by_account["year_week"].str.split("-", expand=True).astype(int)
)

# Create a shifted 'year_week' column for detecting consecutive weeks
weekly_data_by_account["prev_year_week"] = weekly_data_by_account.groupby("profile_id")[
    "year_week"
].shift(1)


# Function to check if the previous week is exactly lag weeks before
def is_exact_lag_vectorized(df, lag):
    prev_week = df.groupby("profile_id")["week"].shift(lag)
    prev_year = df.groupby("profile_id")["year"].shift(lag)

    # Adjust for year transition
    expected_week = df["week"] - lag
    expected_year = df["year"]

    # Handle cases where week becomes negative (previous year)
    expected_year[expected_week <= 0] -= 1
    expected_week[expected_week <= 0] += 52  # Assuming a 52-week year

    return (prev_year == expected_year) & (prev_week == expected_week)


# Create lag features with vectorized checks
for lag in lags:
    for feature in lag_features:
        if feature in weekly_data_by_account.columns:
            lag_col_name = f"{feature}_lag_{lag}"
            weekly_data_by_account[lag_col_name] = weekly_data_by_account.groupby(
                "profile_id"
            )[feature].shift(lag)

            # Set lag values to 0 if the previous week is not consecutive
            weekly_data_by_account.loc[
                ~is_exact_lag_vectorized(weekly_data_by_account, lag), lag_col_name
            ] = 0
```

```python
# Ensure engagement column exists before creating future targets
if "engagement_per_post" in weekly_data_by_account.columns:
    weekly_data_by_account["engagement_future_7d"] = weekly_data_by_account.groupby(
        "profile_id"
    )["engagement_per_post"].shift(-1)
    weekly_data_by_account["engagement_future_14d"] = weekly_data_by_account.groupby(
        "profile_id"
    )["engagement_per_post"].shift(-2)
    weekly_data_by_account["engagement_future_21d"] = weekly_data_by_account.groupby(
        "profile_id"
    )["engagement_per_post"].shift(-3)
    weekly_data_by_account["engagement_future_28d"] = weekly_data_by_account.groupby(
        "profile_id"
    )["engagement_per_post"].shift(-4)
else:
    raise ValueError(
        "'engagement_per_post' column not found in dataset. Cannot create target variables."
    )

# Fill NaNs with 0 in the specified target columns
weekly_data_by_account.fillna(
    {
        "engagement_future_7d": 0,
        "engagement_future_14d": 0,
        "engagement_future_21d": 0,
        "engagement_future_28d": 0,
    },
    inplace=True,
)


# Select features and target
original_features = ["followers", "reach", "impressions", "engagement"]
additional_features = ["monthly_average_engagement_per_post"]
features = (
    [
        col
        for col in weekly_data_by_account.columns
        if col.startswith("engagement_lag")
        or col.startswith("followers_lag")
        or col.startswith("reach_lag")
        or col.startswith("impressions_lag")
        or col.startswith("likes_lag")
        or col.startswith("comments_lag")
    ]
    + original_features
    + additional_features
)

# Ensure selected features exist in the dataset
available_features = weekly_data_by_account.columns.tolist()
missing_features = [col for col in features if col not in available_features]
if missing_features:
    print(
        f"Warning: The following expected features are missing and will be removed: {missing_features}"
    )
    features = [col for col in features if col in available_features]

# Ensure target variable exists
if "engagement_future_7d" not in weekly_data_by_account.columns:
    raise ValueError("Target column 'engagement_future_7d' not found in dataset.")
```

## Import Models

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import joblib  # For saving models
```

```python
import pickle
import os

targets = [
    "engagement_future_7d",
    "engagement_future_14d",
    "engagement_future_21d",
    "engagement_future_28d",
]

metrics = {}

# Path to the folder containing the models
models_folder_path = "models"

# Load all models from the folder
models = {}
for filename in os.listdir(models_folder_path):
    if filename.endswith(".pkl"):  # Ensure to load only pickle files
        model_path = os.path.join(models_folder_path, filename)
        with open(model_path, "rb") as file:
            # The key for the model is set as the filename without the extension
            loaded_object = pickle.load(file)
            models[filename[:-4]] = loaded_object
            print(f"Loaded {filename}: Type = {type(loaded_object)}")


models.pop("scaler", None)  # Remove scaler if present

# Make predictions with each model
predictions = {}
for model_name, model in models.items():
    predictions[model_name] = model.predict(X_test)

# print(predictions)
print(models)
```

## Make Predictions

```python
# Dictionary to store predictions
predictions = {}

# Evaluate each model
for target in targets:
    model_name = f"{target}_model"
    model = models[model_name]

    # Make predictions
    y_pred = model.predict(X_test)
    # y_pred = model.predict(X_test_scaled)
    predictions[target] = y_pred  # Store predictions

    # Calculate evaluation metrics
    mae = mean_absolute_error(y_test[target], y_pred)
    mse = mean_squared_error(y_test[target], y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test[target], y_pred)

    # Store metrics
    metrics[target] = {"MAE": mae, "MSE": mse, "RMSE": rmse, "RÂ²": r2}

    # Print results
    print(f"\nEvaluation for {target}:")
    print(f"MAE: {mae:.4f}")
    print(f"MSE: {mse:.4f}")
    print(f"RMSE: {rmse:.4f}")
    print(f"RÂ²: {r2:.4f}")

print("\nAll predictions and evaluations completed.")
```
